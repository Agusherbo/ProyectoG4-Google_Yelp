```python
# Inicializar el analizador de sentimientos de NLTK
sid= SentimentIntensityAnalyzer()
```

```python
#Prueba
resultado=sid.polarity_scores("it was a horrible experience")
resultado
```

```python
#Analisis de sentimiento para el archivo review

# Retrona 0 si es una reseña negativa
# Retrona 1 si es una reseña nula o neutra
# Retrona 2 si es una reseña positiva

def analyze_sentiment(text):
    if text is None or pd.isna(text):
        return 1 
    else:
        
        sentiment_score = sid.polarity_scores(text)['compound']
        if sentiment_score < -0.05:
            return 0  
        elif sentiment_score > 0.05:
            return 2  
        else:
            return 1  

dfreview['sentiment_analysis'] = dfreview['text'].apply(analyze_sentiment)
```

```python
#Se hace un procesamiento del lenguaje natural con el objetivo de encontrar las palabras mas repetidas en todas las reseñas
todos_los_textos = ' '.join(texto for texto in tacobell['text'] if texto is not None)

# Función para obtener las 10 palabras más repetidas en un texto
def obtener_palabras_mas_repetidas(texto, num_palabras=40):
    stop_palabras = set(stopwords.words('spanish'))
    palabras = word_tokenize(texto.lower())
    words_filtered = [word for word in palabras if word.isalnum() and word not in stop_palabras]
    contador_palabra = Counter(words_filtered)
    palabras_mas_comunes = contador_palabra.most_common(num_palabras)
    return palabras_mas_comunes

# Obtener las 10 palabras más repetidas en el texto combinado
palabras_mas_repetidas = obtener_palabras_mas_repetidas(todos_los_textos)

```
